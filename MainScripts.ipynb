```python
import cv2
import os
import shutil

# Define paths
source_folder = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\photos'  # Source folder containing the images
faces_folder = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\ketPics'        # Destination folder for images with faces
no_faces_folder = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\non-ketPics'  # Destination folder for images without faces

# Create destination folders if they don't exist
os.makedirs(faces_folder, exist_ok=True)
os.makedirs(no_faces_folder, exist_ok=True)

# Load pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Function to detect faces in an image
def detect_faces(image_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    return len(faces) > 0  # Returns True if faces are detected

# Iterate over each file in the source folder
for filename in os.listdir(source_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Check if the file is an image
        image_path = os.path.join(source_folder, filename)
        
        # Detect faces
        if detect_faces(image_path):
            # If faces are detected, move to the faces folder
            shutil.move(image_path, os.path.join(faces_folder, filename))
            print(f"Moved {filename} to {faces_folder}")
        else:
            # If no faces are detected, move to the no_faces folder
#             shutil.move(image_path, os.path.join(no_faces_folder, filename))
            print(f"Moved {filename} to {no_faces_folder}")

print("Image sorting complete.")
```

```python
import os
import shutil
from deepface import DeepFace

# Define paths
source_folder = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\ketPics'  # Folder containing images to search through
target_folder = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\ketPics-ket'       # Folder to move matching images into
reference_image = r'C:\Users\Kevin\Downloads\ketakitamhankar_2299385460297370\ketPics-ket\101665948_546536772697769_4500232510449385472_n_546536762697770.jpg'  # Reference image of the person you're looking for

# Create the target folder if it doesn't exist
os.makedirs(target_folder, exist_ok=True)

# Function to verify if a person's face is present in an image
def is_target_person(image_path, reference_image, model_name="Facenet"):
    try:
        result = DeepFace.verify(img1_path=reference_image, img2_path=image_path, model_name="Facenet", enforce_detection=False)
        return result['verified']
    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")
        return False

# Iterate over images in the source folder
for filename in os.listdir(source_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Only process image files
        image_path = os.path.join(source_folder, filename)
        
        # Check if the target person is in the image
        if is_target_person(image_path, reference_image):
            # Move the image to the target folder if a match is found
            shutil.move(image_path, os.path.join(target_folder, filename))
            print(f"Moved {filename} to {target_folder}")
        else:
            print(f"No match found for {filename}")

print("Image processing complete.")
```
